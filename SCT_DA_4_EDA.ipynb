```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv", sep=";")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Income"], bins=30)
plt.title("Income Distribution")
plt.show()
```

```python
df["Age"] = 2026 - df["Year_Birth"]

plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
products = ["MntWines","MntFruits","MntMeatProducts",
            "MntFishProducts","MntSweetProducts","MntGoldProds"]

df[products].sum().plot(kind="bar")
plt.title("Total Spending by Product")
plt.show()
```

```python
campaigns = ["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3",
             "AcceptedCmp4","AcceptedCmp5","Response"]

df[campaigns].sum().plot(kind="bar")
plt.title("Campaign Acceptance")
plt.show()
```
